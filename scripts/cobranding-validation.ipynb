```r
getRegACQ = function (write = FALSE) {
    
    acq = setACQYML() %>%
        str_match_all(
            pattern = regex(
                '\\n*\\s*(\\w*):\\s*#\\s*(.*)\\n\\s*(?:id:\\s(\\d*))?(?:\\n\\s*co_branding:(?:\\n\\s*enabled:\\s(.*))?(?:\\n\\s*logo_image:\\s\\\"(.*)\\\")?(?:\\n\\s*logo_line:\\s\\\"(.*)\\\")?(?:\\n\\s*tag_line:\\s\\\"(.*)\\\")?)?'
            )
        ) %>% 
        .[[1]] %>%
        as.data.frame(
            stringsAsFactors = FALSE
        ) %>% 
        transmute(
            raw = V1 %>% str_trim('both'),
            short.name = V2 %>% str_trim('both'),
            name = V3 %>% str_trim('both'),
            partner.id = V4 %>% str_trim('both') %>% as.integer(),
            cobranding.enabled = V5 %>% str_trim('both') %>% replace_na('false') %>% as.logical(),
            cobranding.logo = V6 %>% str_trim('both') %>% replace_na(''),
            cobranding.logo.line = V7 %>% str_trim('both') %>% replace_na(''),
            cobranding.tag.line = V8 %>% str_trim('both') %>% replace_na('')
        ) %>% 
        filter(
            cobranding.enabled
        )
    
    if (write)
        acq %>% write.csv('..\\docs\\cobranded.csv')
    
    return(acq)
    
}
```

### ITA Via Scrape
https://www.discover.opploans.com/*

```r
getScrape = function (
    admethods.df = read_csv("..\\data\\other-partners-admethod-2020-02-25.csv"),
    short.names.df = read_csv("..\\data\\partner-short-names.csv"),
    write = TRUE
) {
    
    ##  This gets cobranding details on reg and interstitial status
    
    executeRegScrape = function (ptr.num) {

        html = read_html(
            paste0(
                "https://www.opploans.com/ptr/",
                ptr.num,
                "/lead/",
                runif(
                    n = 1,
                    min = 0,
                    max = 9999
                ) %>%
                str_pad(
                    width = 4,
                    side = "left",
                    pad = "0"
                ) %>% as.numeric() %>% floor()
            )
        ) %>% 
        html_text()

        has.logo = html %>% 
            str_match_all(
                regex(
                    "(src\\s*=\\s*\\\".*?\\.png\\\")"
                )
            ) %>%
            .[[1]] %>%
            as.data.frame(
                stringsAsFactors = FALSE
            ) %>%
            select(
                V2
            ) %>%
            filter(
                !V2 %>% str_detect(
                    "/assets/sample-check-95a2a5cc3e960af25251b199268e980900a7450cf16707f4eaaf3307668cb962.png"
                )
            ) %>% nrow() > 0

        has.logoline = html %>% 
            str_match_all(
                regex(
                    '(<h1 class="logo-line">.*?</h1>)'
                )
            ) %>% 
            .[[1]] %>% 
            as.data.frame(
                stringsAsFactors = FALSE
            ) %>% 
            select(
                V2
            ) %>% nrow() > 0

        has.interstitial = html %>% 
            str_match_all(
                regex(
                    '"show_partner_landing_takeover":(true|false)'
                )
            ) %>% 
            .[[1]] %>% 
            as.data.frame(
                stringsAsFactors = FALSE
            ) %>% 
            transmute(
                V2 = V2 %>% as.logical()
            ) %>% .$V2

        df = data.frame(
            element.name = c('logo', 'logoline', 'interstitial'),
            element.has = c(has.logo, has.logoline, has.interstitial),
            stringsAsFactors = FALSE
        )

        return(df)


    }
    executeITAScrape = function (ptr.short.name) {

        html = try({
            read_html(
                paste0(
                    "http://discover.opploans.com/",
                    ptr.short.name
                )
            ) %>% 
            html_text()
        })

        html %>% class() == 'character'

    }
    
    scrape = admethods.df %>%
        left_join(
            short.names.df,
            by = 'admethod'
        ) %>% 
        mutate(
            cobranding = external_id %>%
                map(
                    executeRegScrape
                ),

            cobranding.logo = cobranding %>%
                map(
                    .f = function (x) {
                        x %>% filter(element.name == 'logo') %>% .$element.has
                    }
                ) %>% as.logical(),

            cobranding.logo.line = cobranding %>%
                map(
                    .f = function (x) {
                        x %>% filter(element.name == 'logoline') %>% .$element.has
                    }
                ) %>% as.logical(),

            interstitial = cobranding %>%
                map(
                    .f = function (x) {
                        x %>% filter(element.name == 'interstitial') %>% .$element.has
                    }
                ) %>% as.logical(),
            
            ita = short_name %>% 
                map(
                    executeITAScrape
                ) %>%
                as.logical()
            
        ) %>% 
        select(
            -cobranding
        ) %>% 
        arrange(
            admethod
        )
    
    if (write)
        scrape %>% write.csv('..\\docs\\partner-experience-scrape.csv')
    
    return(scrape)
    
}
```
